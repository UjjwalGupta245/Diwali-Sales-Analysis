```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

**State**

```python
df.columns
```

```python
# Total number of order from the top 10 state
df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set({'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
# state vs amount
df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(data=state_amount,x='State',y='Amount')
for bars in ax.containers:
  ax.bar_label(bars)
```

we use both the orders and amount as we can see in the first graph kerala is present and in the sencond graph it is not this shows that the ordering number in kerala are high but the amount spent on ordering is low
Uttar Pradesh , Maharastra , Karnatak are top three


**Marital Status**

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Marital_status_ gender vs amount
sales_status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
ax=sns.barplot(data=sales_status,x='Marital_Status',y='Amount',hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

the most amount spent is by the married women and than by the female womens that are not married.

**Occupation**

```python
sns.set(rc={'figure.figsize':(22,5)})
ax=sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_occupation,x='Occupation',y='Amount')
```

The most amount spent by the people working in it sector , healthcare , aviation.


**Product category**

```python
sns.set(rc={'figure.figsize':(30,5)})
ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# product vs amount
sales_product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_product,x='Product_Category',y='Amount')
```

In the two graphs we can clearly see that the most number of orders are of cloths but when we compare it with the second graph we can clearly see that the most amount of money is spent on food .

```python

sales_id=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(data=sales_id,x='Product_ID',y='Orders')
```
